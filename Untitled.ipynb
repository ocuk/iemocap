```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from modules import iemocap
```

## Reading data

```python
label_dir = 'D:\\IEMOCAP_full_release\\'
train_sessions = [1, 2, 3, 4]
data = iemocap.read_labels(label_dir, include_self_evaluation=False, apply_mapping=True)
```

## Dialog length statistics 

```python
grp = data.groupby(level=[0])

turns = []
for dialog, df in grp:
    turns.append(len(df))

plt.title('Number of turns per dialog')
plt.hist(turns, 20)
plt.xlabel('Количество высказываний в диалоге')
plt.ylabel('Количество диалогов')
plt.show()

print('Number of dialogs: ', len(data.index.get_level_values(0).unique()))
print('Number of turns: ', len(data.index.get_level_values(1)))
print()
print('Average: ', np.int32(np.round(np.mean(turns))))
print('Min: ', np.min(turns))
print('Max: ', np.max(turns))
```

## Valence 

```python
valence = data['valence']

f, (ax1, ax2) = plt.subplots(1, 2)


low = 2.3
high = 3.5

valence_clusters = [(lambda val: 0 if val <= low else (2 if val >= high else 1))(val) for val in valence]

ax1.hist(valence, 10)
ax1.set_title('Original values')
ax2.hist(valence_clusters, 3)
ax2.set_title('Clustered values')
plt.show
```

## Activation

```python
activation = data['activation']

f, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(activation, 10)
ax1.set_title('Original values')

low = 2.5
high = 3.5

activation_clusters = [(lambda val: 0 if val <= low else (2 if val >= high else 1))(val) for val in activation]

ax2.hist(activation_clusters, 3)
ax2.set_title('Clustered values')
plt.show
```
